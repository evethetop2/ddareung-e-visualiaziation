## 대여일 데이터 처리 - 월, 요일, 시간 분리

### 대여일, 시간 분리

```python
# 대여일로 대여 요일구하여 컬럼 추가하기
from datetime import datetime

# 대여일 data type을 datetime으로 변경
df['대여일시'] = pd.to_datetime(df['대여일시'])

# dt.weekday 함수로 대여일 요일 변환, 컬럼 추가(Monday=0, Sunday=6)
df['대여 월'] = df['대여일시'].dt.month
df['대여 시간'] = df['대여일시'].dt.hour
df['대여 요일'] = df['대여일시'].dt.weekday
```

```python
df.head()
```

## 강남/강서 데이터 분리 (월, 요일)


```python
# 강남구청 kn_df
kn_df = df[df['대여 대여소명']=='강남구청'].copy().reset_index(drop=True)
kn_df.head()
```

```python
# 요일, 월 별 이용자 수 df 만들기
kn_week = pd.DataFrame(kn_df['대여 요일'].value_counts().sort_index())
kn_week = kn_week.rename(columns={'대여 요일':'강남구청'})

kn_month = pd.DataFrame(kn_df['대여 월'].value_counts().sort_index())
kn_month = kn_month.rename(columns={'대여 월':'강남구청'})
```

```python
# 강서구청 ks
ks_df = df.loc[df['대여 대여소명'] == '강서구청'].copy().reset_index(drop=True)
ks_df.head()
```

```python
ks_week = pd.DataFrame(ks_df['대여 요일'].value_counts().sort_index())
ks_week = ks_week.rename(columns={'대여 요일':'강서구청'})

ks_month = pd.DataFrame(ks_df['대여 월'].value_counts().sort_index())
ks_month = ks_month.rename(columns={'대여 월':'강서구청'})
```

```python
# 시간별 이용자 수 리스트로 만들기 함수
def hour_sum(wh):

  hours = [0 for a in range(24)]

  for i in range(24):
    hours[i] = len(wh.loc[wh['대여일시'].dt.hour == i])

  return hours
```

```python
ks_hours = hour_sum(ks_df)
kn_hours = hour_sum(kn_df) 
total_hours = hour_sum(df)
```
